```python
d=pd.read_csv(r'C:\Users\Admin\Desktop\Kerfegar\Analytics&ML\Projects\SPARK GRIP Internship\Task5\deliveries.csv')
d
```

```python
d.isna().sum()
d.dropna(axis=1,inplace=True)
```

```python
runs_per_team=[]
teams=d['batting_team'].unique()
teams
for i in teams:
    s=d[d['batting_team']==i]['total_runs'].sum()
    runs_per_team.append(s)
tr=pd.DataFrame({'team':teams,'runs':runs_per_team})
tr
```

```python
plt.figure(figsize=[30,8])
plt.bar(tr['team'],tr['runs'], align='center')
plt.xticks(tr['team'])
plt.show()
```
