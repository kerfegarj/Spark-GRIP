```python
#Splitting the data into feature and target variables
y = df['Species']
x = df.drop(['Species'],axis=1)
```

```python
#Splitting the datta into training and testing
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x, y, test_size = 0.2)
```

```python
#importing the model
from sklearn.tree import DecisionTreeClassifier
sc=[]
#Finding the optimum max depth of the decision tree
for i in range(1,200):
    model=DecisionTreeClassifier(max_depth=i)
    model.fit(xtrain,ytrain)
    sc.append(model.score(xtest,ytest))
```

```python
#Plotting for optimum max depth of the decision tree
plt.plot(range(1,200),sc)
plt.xlabel("Max Depth")
plt.ylabel("Score")
```

```python
#Fiting the decision tree for a max depth of 3(optimum max depth)
model=DecisionTreeClassifier(max_depth=3)
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
#Generating predictions
ypreds=model.predict(xtest)
#Comparing the actual and predicted values
comp=pd.DataFrame({"Actual":ytest,"Predicted":ytest})
comp
```

```python
#Deriving the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,ypreds)
```

```python
#Visualising the decision tree
from sklearn import tree
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,10))
tree.plot_tree(model,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'] ,
               class_names = df['Species'].unique(), filled=True)
plt.show()
```
